```python
#Importing required moduls
import turtle
import time
import random

#game setting initilization
delay =0.1
gamescore =0
highest_score = 0
 
#Window screen of game
ws = turtle.Screen()
ws.title('Snake_game @TokyoEdTech')
ws.bgcolor("plum")
ws.setup(width=600,height =600)
ws.tracer(0)


#Defining Snake head characters
head = turtle.Turtle()
head.shape("square")
head.color("lightblue")
head.penup()
head.goto(0,0)
head.direction = "Stop"


#Defining snake food characters
food = turtle.Turtle()
colors = random.choice (["yellow","red","blue"])
shapes = random.choice (["triangle","circle"])
food.speed(0)
food.shape(shapes)
food.color(colors)
food.penup()
food.goto(0,100)

#Empty list
segments =[]

#defining background area of window screen
pen = turtle.Turtle()
pen.speed(0)
pen.shape("square")
pen.color("white")
pen.penup()
pen.hideturtle()
pen.goto(0,260)
pen.write("score:{} highest_score :{}".format(score,highest_score), align = "center",
             font = ("candara",24, "bold"))


#Creating functions to define directions of snake head
def go_up():
    if head.direction != "down":
        head.direction = "up"

def go_down():
    if head.direction != "up":
        head.direction = "down"
        
def go_left():
    if head.direction != "right":
        head.direction = "left"
        
def go_right():
    if head.direction != "left":
        head.direction = "right"

#creating functions to decide snake head direction as per last movement
def snake_move():
    if head.direction == "up":
        y = head.ycor()
        head.sety(y+20)

    if head.direction == "down":
        y = head.ycor()
        head.sety(y-20)

    if head.direction == "left":
        x = head.xcor()
        head.setx(x-20)

    if head.direction == "right":
        x = head.xcor()
        head.setx(x+20)

#defining keyboard buttond to move/play the game
ws.listen()
ws.onkeypress(go_up,"Up")
ws.onkeypress(go_down,"Down")
ws.onkeypress(go_left,"Left") 
ws.onkeypress(go_right,"Right")

#main game loop
while True:
    ws.update()
    
    #creating if condition to check collisions with the border
    if head.xcor()>290 or head.xcor()<-290 or head.ycor()>290 or head.ycor()<-290:
        time.sleep(1)
        head.goto(0,0)
        head.direction='Stop'
     
        #hiding the segments
        for segment in segments:
            segment.goto(1000,1000)
        
        #clearing segments list
        segments.clear()
        
        #reset the score 
        score = 0
        
        #reset the delay
        delay = 0.1
            
        pen.clear()
        pen.write("score:{} highest_score :{}".format(score,highest_score), align = "center",
             font = ("candara",24, "bold"))

    #creating if condition to check collisions with the food
    if head.distance(food)<20:
        x = random.randint(-270,270)
        y = random.randint(-270,270)
        food.goto(x,y)

        #adding bonus segment
        bonus_segment = turtle.Turtle()
        bonus_segment.speed(0)
        bonus_segment.shape("square")
        bonus_segment.color("grey")
        bonus_segment.penup()
        
        segments.append(bonus_segment)
        
        delay -=0.001
        
        score +=10
        
        #if score > highest_score:
            #highest_score = score
            
        pen.clear()
        pen.write("score:{} highest_score :{}".format(score,highest_score), align = "center",
             font = ("candara",24, "bold"))

    #for loop to move end segments in reverse order
    for index in range(len(segments)-1,0,-1):
        x=segments[index-1].xcor()
        y=segments[index-1].ycor()
        segments[index].goto(x,y)

    #Move segments 0 to head position
    if len(segments)>0:
            x=head.xcor()
            y= head.ycor()
            segments[0].goto(x,y)
            
    snake_move()
    
    #creating for loop to check head collision with body segments
    for segment in segments:
        if segment.distance(head) < 20:
            time.sleep(1)
            head.goto(0,0)
            head.direction = "stop"
        
            # Hide the segments
            for segment in segments:
                segment.goto(1000, 1000)
        
            # Clear the segments list
            segments.clear()

            # Reset the score
            score = 0

            # Reset the delay
            delay = 0.1
        
            # Update the score display
            pen.clear()
            pen.write("Score: {}  High Score: {}".format(score, high_score), align="center", font=("Courier", 24, "normal"))

    time.sleep(delay)
    
ws.mainloop()
```
